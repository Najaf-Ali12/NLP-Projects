```python
# Dividing in X and y
X=data['Text']
y=data['Language']
```

# Text Preprocessing

```python
# Encoding the y
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
```

```python
# Applying nlp on X
import re
data_list = []
for text in X:
    text = re.sub(r'[!@#$(),\n"%^*?\:;~`0-9]', ' ', text)
    text = re.sub(r'[[]]', ' ', text)
    text = text.lower()
    data_list.append(text)

data_list
```

```python
# Applying bag of words
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
data_list=cv.fit_transform(data_list)
data_list
```

```python
i=0
for each in data_list:
    print(each)
    i+=1
    if i>15:
        break
```

```python
# Getting some words at particular position
feature_names=cv.get_feature_names_out()
print("At position 24351 the word is ",feature_names[24351])
print("At position 16409 the word is ",feature_names[16409])
```

```python
X=data_list.toarray()
X=pd.DataFrame(X)
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.8,random_state=42)
```

# Machine Learning model (MultinomialNB)

```python
# Importing the machine learning model
from sklearn.naive_bayes import MultinomialNB  # Naive Bayes algorithm is for classification only
nb=MultinomialNB()
```

```python
# Training the model
nb.fit(X_train,y_train)
```

```python
# Getting predictions from the model
y_pred=nb.predict(X_test)
```

```python
# Getting metrics
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
ac=accuracy_score(y_pred,y_test)
cm=confusion_matrix(y_pred,y_test)
cr=classification_report(y_pred,y_test)
```

```python
# Accuracy
print("Accuracy: ",ac)
```

```python
# Classification report
print(cr)
```

```python
# Confusion_matrix
print("Confusion Matrix")
print(cm)
```

```python
# Visualizing the confusion_matrix
import seaborn as sns
plt.figure(figsize=(12,10))
sns.heatmap(cm,annot=True)
plt.show()
```
